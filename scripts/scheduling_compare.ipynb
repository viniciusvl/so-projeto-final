```python
import pandas
```

```python
# Coletando os dados dos arquivos de log

with open('../out/com1.out', 'r') as f:
    com1_data = f.readlines()

data_benchmark = {
    'PID': [],
    'Pit Tick': [],
    'Event ID': [],
    'Context ID': []
}

for line in com1_data:
    if line.startswith('[STAT]'):
        parts = line.strip().split(' ')[1].split(',')
        data_benchmark['PID'].append(parts[0])
        data_benchmark['Pit Tick'].append(parts[1])
        data_benchmark['Event ID'].append(parts[2])
        data_benchmark['Context ID'].append(parts[3])

# Criando um DataFrame com os dados coletados
df_benchmark = pandas.DataFrame(data_benchmark)

# Exibindo o DataFrame
print(df_benchmark)

# Salvando o DataFrame em um arquivo CSV
df_benchmark.to_csv('benchmark_data.csv', index=False)
```
